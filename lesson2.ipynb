```python
# Создайте одномерный массив c, составленный из последовательности 16-ти случайных равномерно распределенных целых чисел 
# от 0 до 16 невключительно. 
c = np.random.randint(0, 16, 16)
c
```

```python
# Поменяйте его форму так, чтобы получилась квадратная матрица C.
C = c.reshape(4, 4)
C
```

```python
# Получите матрицу D, поэлементно прибавив матрицу B из предыдущего вопроса к матрице C, умноженной на 10.
D = np.add(B, C*10)
D
```

```python
# Вычислите определитель
np.linalg.det(D)
```

```python
# ранг
np.linalg.matrix_rank(D)
```

```python
# и обратную матрицу D_inv для D.
D_inv = np.linalg.inv(D)
D_inv
```

```python
#Приравняйте к нулю отрицательные числа в матрице D_inv, а положительные - к единице. Убедитесь, что в матрице D_inv 
# остались только нули и единицы.
D_inv = np.where(D_inv>0, D_inv, 0)
D_inv = np.where(D_inv<=0, D_inv, 1)
D_inv
```

```python
# С помощью функции numpy.where, используя матрицу D_inv в качестве маски, а матрицы B и C - в качестве источников данных, 
# получите матрицу E размером 4x4.  Элементы матрицы E, для которых соответствующий элемент  матрицы D_inv равен 1, 
# должны быть равны соответствующему элементу матрицы B, а элементы матрицы E, для которых соответствующий элемент 
# матрицы D_inv равен 0, должны быть равны соответствующему элементу матрицы C.
E = np.where(D_inv==1, B, C)
E
```
